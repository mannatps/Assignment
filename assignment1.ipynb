# Question 1 
#### Part 1 - File method

```python
import nltk
import os
path = 'C:/Users/Immortal Blue/PycharmProjects/udacity/txt_corpa'
fns = os.listdir(path)
print(fns, "\n\n")
content = list()
cx = []
for i in fns:
    file = open(path + '/' + i, 'r', encoding="utf8")
    text = file.read()
    content.append([text])
    cx.append(text)
    file.close()

print(content)
```

#### Part 2 - PlaintextCorpusReader

```python
from nltk.corpus import PlaintextCorpusReader
path = 'C:/Users/Immortal Blue/PycharmProjects/udacity/txt_corpa'
dataset = PlaintextCorpusReader(path, '.*')
raw_text_list = []
for i in range(len(dataset.fileids())):
    raw_text_list.extend([[dataset.raw(fileids=dataset.fileids()[i])]])

print(raw_text_list)
```

# Question 2

```python
# normalizing
lower_corpus = list()
for corpus in raw_text_list:
    lower_corpus.append(' '.join([word.lower() for word in corpus[0].split()]))

#print(lower_corpus)
alpha_corpus = []
for i in lower_corpus:
    alpha_corpus.append(' '.join([word for word in i.split() if word.isalpha()]))
print("\n\n\n", alpha_corpus)
```

```python
# tokenization
tokenized_corpus = []
for cor in alpha_corpus:
    tokenized_corpus.append(nltk.word_tokenize(cor))
print(tokenized_corpus)
```

```python
# Stopword Removal
from nltk.corpus import stopwords
stop=stopwords.words(fileids='english')
tokenized_corpus_nostop = []
for token in tokenized_corpus:
    tokenized_corpus_nostop.append([word for word in token if word not in stop])

print(tokenized_corpus_nostop)
```

```python
# Stemming
from nltk.stem import PorterStemmer
porstem = PorterStemmer()
tokenized_corpus_nostop_stem=[]
for cor in tokenized_corpus_nostop:
    tokenized_corpus_nostop_stem.append([porstem.stem(word) for word in cor])
    
print(tokenized_corpus_nostop_stem)
```

#### Part 1

```python
# CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
#for i in content:
    
vec = CountVectorizer()
vec2 = TfidfVectorizer()
X = vec.fit_transform(cx)
Y = vec2.fit_transform(cx)
print("TfidfVectorizer Shape : ", Y.shape)
print("CountVectorizer : \n", vec.get_feature_names())
print(X.toarray())
print("\n\n\n\n")
```

#### Part 2

```python
import pandas as pd

listofwords=[]
for i in tokenized_corpus_nostop:
    listofwords.extend(i)
uniqueWords = set(listofwords)
num = []
for i in tokenized_corpus_nostop:
    numOfWords = dict.fromkeys(uniqueWords, 0)
    for w in i:
        numOfWords[w] += 1
    num.append(numOfWords)

def computeTF(wordDict, bagOfWords):
    tfDict = {}
    bagOfWordsCount = len(bagOfWords)
    for word, count in wordDict.items():
        tfDict[word] = count / float(bagOfWordsCount)
    return tfDict

tf = []
for i in range(len(num)):
    tfDict = computeTF(num[i], tokenized_corpus_nostop[i])
    tf.append(tfDict)

def computeIDF(documents):
    import math
    N = len(documents)
    
    idfDict = dict.fromkeys(documents[0].keys(), 0)
    for document in documents:
        for word, val in document.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        idfDict[word] = math.log(N / float(val))
    return idfDict

idfs = computeIDF(num)

def computeTFIDF(tfBagOfWords, idfs):
    tfidf = {}
    for word, val in tfBagOfWords.items():
        tfidf[word] = val * idfs[word]
    return tfidf
tfidfs = []
for i in range(len(num)):
    tfidf = computeTF(tf[i], idfs)
    tfidfs.append(tfidf)

    
df = pd.DataFrame(tfidfs)
print(df)
```
