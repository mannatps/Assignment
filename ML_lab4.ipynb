```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/home/USA_Housing.csv')
df.head()
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

**Q1**


```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_scale = sc.fit_transform(x)
x_scale = np.insert(x_scale,0,values=1,axis=1)
```

```python
from sklearn.model_selection import RepeatedKFold
from sklearn.metrics import r2_score 
kf = RepeatedKFold(n_splits = 5, n_repeats = 1, random_state = None) 
betas = []
errors = []
r2 = []
for train_index,test_index in kf.split(x_scale):
  print("Train:", train_index, "Validation:",test_index)
  x_train,x_test = x_scale[train_index],x_scale[test_index] 
  y_train,y_test = y[train_index],y[test_index]
  A = x_train.T.dot(x_train)
  B = np.linalg.inv(A)
  C = B.dot(x_train.T)   
  beta = C.dot(y_train)
  betas.append(beta)
  y_pred = x_test.dot(beta)
  r2.append(r2_score(y_test, y_pred))
  error = y_test-y_pred
  square_error = np.power(error,2)
  sum_square_error = np.sum(square_error)
  mse = sum_square_error/len(y_pred)
  errors.append(mse)
```

```python
r2
```

```python
max_index = r2.index(max(r2))
```

```python
betas[max_index]
```

```python
best_beta=betas[max_index]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scale, y, test_size = 0.3, random_state = 1)
y_pred = x_test.dot(best_beta)
print(y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scale, y, test_size = 0.3, random_state = 1)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size = 0.2, random_state = 1)
```

```python
learning_rates = [0.001, 0.01, 0.1, 1]
iterations = 1000
n = len(x_train)
betas=[]
```

```python
for learning in learning_rates:
  beta = np.zeros(6)
  for i in range(iterations):
    x0_gradient=0
    x1_gradient=0
    x2_gradient=0
    x3_gradient=0
    x4_gradient=0
    x5_gradient=0
    for j in range(n):
      a = x_train[j,0]
      b = x_train[j,1]
      c = x_train[j,2]
      d = x_train[j,3]
      e = x_train[j,4]
      f = y_train[j]
      x0_gradient+=(beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)+(beta[4]*d)+(beta[5]*e)-f)
      x1_gradient+=((beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)+(beta[4]*d)+(beta[5]*e)-f)*a)
      x2_gradient+=((beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)+(beta[4]*d)+(beta[5]*e)-f)*b)
      x3_gradient+=((beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)+(beta[4]*d)+(beta[5]*e)-f)*c)
      x4_gradient+=((beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)+(beta[4]*d)+(beta[5]*e)-f)*d)
      x5_gradient+=((beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)+(beta[4]*d)+(beta[5]*e)-f)*e)
    beta[0]=beta[0]-learning/n*x0_gradient
    beta[1]=beta[1]-learning/n*x1_gradient
    beta[2]=beta[2]-learning/n*x2_gradient
    beta[3]=beta[3]-learning/n*x3_gradient
    beta[4]=beta[4]-learning/n*x4_gradient
    beta[5]=beta[5]-learning/n*x5_gradient
  betas.append(beta)
```

```python
betas
```

```python
r2 = []
from sklearn.metrics import r2_score
for beta in betas:
  y_pred_val = x_val.dot(beta)
  r2.append(r2_score(y_val, y_pred_val))
```

```python
max_index = r2.index(max(r2))
optimal_beta = betas[max_index]
y_pred_final = x_test.dot(optimal_beta)
r2_final = r2_score(y_test, y_pred_final)
```
